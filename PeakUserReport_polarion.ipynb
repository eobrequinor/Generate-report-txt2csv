```python
def isNaN(num):
    return num != num

def generate_CSV(data):
    data = pd.read_csv(data, header=None)
    df1 = data[[2,3,4]]
    df1.rename({2: 'current', 3: 'peak', 4: 'limit'}, axis=1, inplace=True)
    df1[['Date','Timestamp']] = data[0].str.split(' ', expand=True)
    df2 = df1[['Date','Timestamp','current', 'peak', 'limit']]

    df2 = df2[df2['peak']==df2['peak']]
    
    #Removes letters from the current/peak/limit cells so there is only a single nummeric value
    df2['current']=df2['current'].str.replace('current:','')
    df2['peak']=df2['peak'].str.replace('peak:','')
    df2['limit']=df2['limit'].str.replace('limit:','')
   
        
    return df2
```

```python
# Folder Path
path = "C:/Users/emil.obrestad/Equinor/PolarionLogs/logs"
df_ferdig = []
  
# Change the directory
os.chdir(path)
# iterate through all file
for file in os.listdir():
    # Check whether file is in text format or not
    if file.endswith(".log"):
        file_path = f"{path}\{file}"
        print(file_path)
        # call read text file function
        df_ferdig.append(generate_CSV(file_path))
        
#Combines different dataframes into a single one
result = pd.concat(df_ferdig)      
```
